```python
# Pemanggilan Data
```

```python
import pandas as pd

data_febi = pd.read_csv('K02_diabetes.csv')
data_febi
```

```python
# Preprocessing Data
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
encode = LabelEncoder()

for col in data_febi:
    if data_febi[col].dtype == 'object':
        data_febi[col] = encode.fit_transform(data_febi[col])
data_febi
```

```python
X = data_febi.drop('diabetes', axis=1)
Y = data_febi['diabetes']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
# Modelling
```

```python
from sklearn.naive_bayes import GaussianNB

febi = GaussianNB()
febi.fit(x_train, y_train)
```

```python
# Analisis Akurasi Model
```

```python
print("Akurasi dari model ini adalah: ", febi.score(x_train, y_train))
```
