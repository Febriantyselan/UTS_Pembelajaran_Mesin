# Pemanggilan Data

```python
import pandas as pd

data_febi = pd.read_csv('R02_rice_field.csv')
data_febi
```


# Preprocessing Data

```python
from sklearn.model_selection import train_test_split

X = data_febi['Production (X)'].values.reshape(-1,1)
Y = data_febi['Harvest Area (Y)'].values

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2)
```

# Modelling

```python
from sklearn.linear_model import LinearRegression

febi = LinearRegression()
febi.fit(x_train, y_train)
```

# Analisis Akurasi Model

```python
print("Akurasi model: ", febi.score(x_train, y_train))
```

# Pengujian model

```python
prediksi = febi.predict(x_test)
prediksi
```

# Visualisasi Data

```python
import matplotlib.pyplot as plt

plt.scatter(x_test, y_test, c='green')
plt.plot(x_test, prediksi, c='red')
plt.xlabel("Production")
plt.ylabel("Harvest Area")
plt.show()
```
